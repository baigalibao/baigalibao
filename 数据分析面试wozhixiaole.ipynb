import numpy as np
from scipy.stats import ttest_ind

dates = ["2024/1/1", "2024/1/2", "2024/1/3", "2024/1/4", "2024/1/5", "2024/1/6", "2024/1/7", "2024/1/8", "2024/1/9", "2024/1/10", "2024/1/11", "2024/1/12", "2024/1/13", "2024/1/14"]
exp_users = [200681, 202322, 203036, 203451, 204281, 204160, 204869, 202663, 200547, 201355, 202521, 203253, 203129, 200701]
ctrl_users = [200762, 201498, 201970, 201345, 203643, 201624, 201665, 203574, 200124, 204372, 203189, 201938, 202661, 202206]
exp_retention = [0.1517, 0.1548, 0.1550, 0.1532, 0.1505, 0.1501, 0.1530, 0.1518, 0.1529, 0.1505, 0.1529, 0.1541, 0.1549, 0.1518]
ctrl_retention = [0.1523, 0.1554, 0.1555, 0.1486, 0.1510, 0.1507, 0.1535, 0.1526, 0.1535, 0.1460, 0.1535, 0.1546, 0.1557, 0.1523]

exp_retained = [round(users * rate) for users, rate in zip(exp_users, exp_retention)]
ctrl_retained = [round(users * rate) for users, rate in zip(ctrl_users, ctrl_retention)]

exp_data = np.concatenate([np.concatenate([np.ones(retained), np.zeros(total - retained)]) for retained, total in zip(exp_retained, exp_users)])
ctrl_data = np.concatenate([np.concatenate([np.ones(retained), np.zeros(total - retained)]) for retained, total in zip(ctrl_retained, ctrl_users)])

t_stat, p_value = ttest_ind(exp_data, ctrl_data, equal_var=False)

print(f'T-statistic: {t_stat}, P-value: {p_value}')

根据计算结果，t统计量约为0.490，p值约为0.624。这意味着在统计学上，实验组与对照组之间的留存率差异不具有显著性（一般认为p值小于0.05时差异显著）。
